```python
spacy_obj=spacy.load('en_core_web_sm')
```

```python
text_doc=spacy_obj(text_data)
```

```python
type(text_doc)
```

```python
for i in text_doc:
    print(i.text)
```

```python
words=[i.text for i in text_doc]
words
```

```python
word_frequency={}
```

```python
#calculate word_frequency
for word in text_doc:
    if word.text.lower() not in stopwords and word.text not in punctuation:
        if word.text not in word_frequency.keys():
            word_frequency[word.text]=1
        else:
            word_frequency[word.text]=word_frequency[word.text]+1
            
word_frequency
```

```python
word_frequency.values()
```

```python
max_frequency=max(word_frequency.values())
max_frequency
```

```python
#Normalization of word freuency
for word in word_frequency.keys():
    word_frequency[word]=word_frequency[word]/max_frequency

print(word_frequency)
```

```python
text_doc.sents
```

```python
sentence_tokens=[sent for sent in text_doc.sents]
sentence_tokens
```

```python
len(sentence_tokens)
```

```python
#calculate sentence scores
```

```python
sentence_scores={}
```

```python
for sentence in sentence_tokens:
    for word in sentence:
        if word.text in word_frequency.keys():
            if word.text in word_frequency.keys():
                sentence_scores[sentence]=word_frequency[word.text]
            
            else:
                print("inside else --- ",word.text)
                sentence_scores[sentence]=sentence_scores[sentence]+word_frequency[word.text]
                #x=x+y
                #x+=y  
            
print(sentence_scores)
```

```python
len(sentence_tokens)
```

```python
summary_length=int(len(sentence_tokens)*0.30)
summary_length
```

```python
summary=nlargest(summary_length,sentence_scores,key=sentence_scores.get)
summary
```

```python
type(summary[0])
```

```python
text=[word.text for word in summary]
text
```

```python
final_summary=" ".join(text)
final_summary
```
